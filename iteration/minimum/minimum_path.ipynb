```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.DataFrame()
df['Path'] = np.arange(50, 1000 + 1, step=50)
df['Minimum'] = np.genfromtxt('../../dataset/1000_iteration/minimum/ch_min_simplified.csv')
df['0'] = np.genfromtxt('../../dataset/1000_iteration/minimum/ch_min_1000avg_0.csv')
df['5'] = np.genfromtxt('../../dataset/1000_iteration/minimum/ch_min_1000avg_5.csv')
df['10'] = np.genfromtxt('../../dataset/1000_iteration/minimum/ch_min_1000avg_10.csv')
df['20'] = np.genfromtxt('../../dataset/1000_iteration/minimum/ch_min_1000avg_20.csv')
df['30'] = np.genfromtxt('../../dataset/1000_iteration/minimum/ch_min_1000avg_30.csv')
```

```python
%matplotlib notebook
# plt.rc('font', size=12, family='Times New Roman', weight='light')

fig, channel = plt.subplots(figsize=(6, 4))

# channel.set_xlabel('Jumlah Jalur')
# channel.set_ylabel('Koefisien Kanal')
# channel.set_title('Koefisien Kanal Batas Bawah')
channel.plot(df['Path'], df['30'], label='30Â°', marker='o', color='black')
channel.plot(df['Path'], df['20'], label='20Â°', marker='+', color='black')
channel.plot(df['Path'], df['10'], label='10Â°', marker='s', color='black')
channel.plot(df['Path'], df['5'], label='5Â°', marker='D', color='black')
channel.plot(df['Path'], df['0'], label='0Â°', marker='x', color='black')
channel.plot(df['Path'], df['Minimum'], label='Min', color='black')
channel.legend(loc=2, prop={'size': 7})
channel.grid(linewidth='1', linestyle=':')

plt.tight_layout()
plt.show()
plt.savefig('../../images/batasbawah_20path.png')
```
